```python
import numpy as np
white_noise = np.random.normal(0,50,500)

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 5))
ax1.plot(white_noise)
ax1.legend(['White noise'], prop={'size': 14})

signal = range(0,500)
ax2.plot(signal)
ax2.legend(["Signal"], prop={'size': 14})

ax3.plot(signal + white_noise)
ax3.legend(["Signal + white noise"], prop={'size': 14})
```
